```python
import numpy as np

def generate_stock_prices(num_datapoints):
    """
    Generates a series of realistic stock prices using a Geometric Brownian Motion model.

    Args:
        num_datapoints: The number of data points (prices) to generate.

    Returns:
        A numpy array of generated stock prices.
    """

    # Initial parameters
    initial_price = 100.0  # Starting stock price
    mu = 0.05              # Mean annual return (drift)
    sigma = 0.20           # Annual volatility
    dt = 1/252             # Time step (daily assuming 252 trading days per year)

    # Generate random log-returns
    log_returns = np.random.normal(loc=(mu - sigma**2 / 2) * dt, 
                                  scale=sigma * np.sqrt(dt), 
                                  size=num_datapoints)

    # Calculate stock prices
    prices = np.exp(np.cumsum(log_returns)) * initial_price

    return prices
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Assuming the 'generate_stock_prices' function from the previous response is defined

# Generate data
num_datapoints = 252
prices = generate_stock_prices(num_datapoints)
dates = np.arange(num_datapoints) 

# Plotting
plt.figure(figsize=(12, 6)) 

for i in range(1, len(prices)):
    color = 'green' if prices[i] > prices[i - 1] else 'red'
    plt.plot([dates[i - 1], dates[i]], [prices[i - 1], prices[i]], color=color, linestyle='-', marker='o', markersize=5)

# Enhancements
plt.title("Simulated Stock Prices")
plt.xlabel("Trading Days")
plt.ylabel("Price ($)")
plt.grid(axis='y', linestyle='--')

# Show the plot
plt.show()
```
